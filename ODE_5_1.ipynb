# Oscilador Armónico Forzado No Amortiguado:

Considerando el desarrolo realizando durante el laboratorio 3, y teniendo en cuenta que
\begin{eqnarray}
\mathbf{m\ddot{x} + kx = 0}   \ \ \ (\ast)
\end{eqnarray}

## Considerando :

\begin{eqnarray*}
u & = & x \hspace{2cm}  v & =& \dot{x} \\
\dot{u} & = & \dot{x} \hspace{2cm}  \dot{v} & =& \ddot{x} = -\frac{k}{m} x \\
\dot{u} & = & v \hspace{2cm}  \dot{v} & =&  -\frac{k}{m} x
\end{eqnarray*}

## Donde:
\begin{eqnarray}
\begin{bmatrix}
 \dot{u} \\
 \dot{v} 
\end{bmatrix}
& = & 
\begin{bmatrix}
0 & 1 \\
-\frac{k}{m} & 0
\end{bmatrix}
\cdot 
\begin{bmatrix}
u \\
v
\end{bmatrix}
\end{eqnarray}

## Llamando:
\begin{eqnarray}
 A & = & 
 \begin{bmatrix}
0 & 1 \\
-\frac{k}{m} & 0
\end{bmatrix}
\ \ \ \hspace{2cm} 
A =
\left \{
      \begin{array}{lccl}
         \vert A \vert & = & r_{1} \cdot r_{2}  \\
         tr(A) &  =  & - (r_{1} + r_{2})  \\ 
      \end{array}
   \right .
\end{eqnarray}

Donde $r_{1}$ y $r_{2}$ son las raíces del polinomio característico que se obtiene de $(\ast)$. De este modo se puede realizar un análisis de la relación entre la traza y el determinante para tener la información de un punto silla, sumidero, fuente o un sistema aleatorio

# Resultados del Laboratorio:
Luego de una descripción analítica realizada en el laboratorio desarrolados, se procede a:
\begin{eqnarray*}
sol_{G} & = & sol_{H} + sol_{P}
\end{eqnarray*}

### Se tiene así, las raíces: $r_{a} = \alpha + \beta i; \ \ \ \ r_{b} = \alpha - \beta \cdot i  \ \  \Longrightarrow \ \ r_{a} = \sqrt{\frac{k}{m}} \cdot i \ \ \ r_{b} = -\sqrt{\frac{k}{m}} \cdot i  $ 

1. Luego de realizar las operaciones necesarias se obtiene: $ sol_{H} = a e^{\alpha t} \cdot \cos(\beta t) + b \cdot e^{\alpha t} \sin{(\beta t)} = a \cos \big( \sqrt{\frac{k}{m}} t \big) + b \sin \big( \sqrt{\frac{k}{m}} t \big)$

2. La solución paricualar para el problema planteado en el laboratorio es:
\begin{eqnarray*}
sol_{P} & = \frac{10}{k_{1}} + \frac{0.1}{k_{1}-\omega^{2}} \cdot \sin{(\omega t)} &
\end{eqnarray*}

```python
import numpy as np 
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-15,15,1000)
w = 3.51
u = [1,2] #y[0]=x and y[1]=v
```

```python
def H(t,u):
    S = [u[1],(-w**2)*u[0]] #Representa el vector  (x*,y*) =(v,-omega^2 * x )
    return S                #Devuelve la solución
Osc = solve_ivp(H, [-15,1000], y0 = u, t_eval = t) 
```

```python
plt.plot(t,Osc.y[0])
plt.ylabel("Posición ")
plt.xlabel("Tiempo ")
plt.title('Oscilador Armónico', fontsize = 20)
```

```python
t = np.linspace(-15,15,1000)
y = [0,1]
gamma = 1
omega_sqr = 100
def sho(t,y):
    solution = (y[1],(-gamma*y[1]-omega_sqr*y[0]))
    return solution
solution = solve_ivp(sho, [-15,1000], y0 = y, t_eval = t)
plt.plot(t,solution.y[0])
plt.ylabel("Position")
plt.xlabel("Time")
plt.title('Relación entre amplitud y Frecuencia', fontsize = 15)
```
