## Resolución de Ecuaciones diferenciales ordinarias utilizando JULIA

### Se realiza la implementación en JULIA para resolver el problema de valor inicial de:

\begin{eqnarray*}
 \left\{ \begin{array}{lcc}
             \dot{y} = -2 \cdot y + 1   \\
             \\ y(0) = \frac{5}{2}
             \end{array}
   \right.
\end{eqnarray*}


```julia
# Describe la ODE
f(y,p,t) = -2*y+1
```

```julia
## Usa el paquete de ODE (A cargo del leector)
f(y,p,t) = -2*y+1 # Nuevamente describe se describe la ODE
y0 = 2.5          # Asigna el valor inicial --- Problema de Cauchy
tspan = (-1.0,0.0) # Inicio del proceso numérico para obtener la solución, ingresado por usuario
prob = ODEProblem(f,y0,tspan) # Indica a JULIA el problema de valor inicial
```

```julia
# Resuelve el PVI para obtener el listado de valores numéricos 
sol = solve(prob)
```

```julia
# Plotea la solución particular
using Plots; gr()
plot(sol)
```
