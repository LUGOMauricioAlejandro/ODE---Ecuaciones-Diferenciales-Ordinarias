# Sistema Lotka-Volterra

### Al siguiente implementación se realiza de acuerdo al capítulo 6.4 de libro de Strogatz -- Rabbits versus sheep -- donde realiza la descripción del sistema clásico LOTKA-VOLTERRA de competición de especies  

### La descripción formal del sistema está comprendido utilizando la notación de Newton:

\begin{eqnarray}
\left \{
\dot{x} = x \cdot (3-x-2y) \atop 
\dot{y} = y \cdot (2-x-y)
\right.
\end{eqnarray}

### Donde $x(t); \ \ \ y(t)$ describen la población de conejos y la población de ovejar respecivamente

### Considerando el sistema en su forma expresión matricial:

\begin{eqnarray*}
A 
&=&
\begin{bmatrix}
\dfrac{\partial \dot{x}}{\partial{x}} & \dfrac{\partial \dot{x}}{\partial{y}} \\
\dfrac{\partial \dot{y}}{\partial{x}} & \dfrac{\partial \dot{y}}{\partial{y}}
\end{bmatrix}
& = &
\begin{bmatrix}
3-2x-2y & -2x \\
-y & 2-x-2y
\end{bmatrix}
\end{eqnarray*}

```julia
function lotka_volterra(ds,s,m,t)
    x,y = s
    a,b,c,d = m
    ds[1] = dx = x*(a -x - b*y)
    ds[2] = dy = y*(-c -x + d*x)
end
```
