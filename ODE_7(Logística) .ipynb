## Ecuación Logística 

### La siguiente implementación está realizada utilizando la sección  2.3 del libro "NONLINEAR DYNAMICS AND CHAOS" de Steven H. Strogatz

### Sugerido por primera vez para describir el crecimiento de las poblaciones humanas por Verhulst en 1838

### El autor la utiliza en el contexto, para la variación de población humana, la relación entre $\dot{\mathbf{N}}$ y $\mathbf{N}$; donde $\delta$ describe la taza de crecimiento y una capacidad de carga $\mathbf{K}$  

## Forma de la ecuación logística 

\begin{eqnarray}
\dot{N}&=& \delta N \Big( 1 - \dfrac{N}{K} \Big)
\end{eqnarray}

## Solución Analítica de la Ecuación Logística:

### La ecuación logpistica tiene por solución analítica la siguiente expresión:

\begin{eqnarray}
N(t) &=& \dfrac{ K P_{0} e^{\delta \cdot t } }{ K + P_{0} \big( e^{\delta \cdot t } - 1 \big) }
\end{eqnarray}

### Donde $P_{0}$ es la población inicial con la cual comienza el análisis de la ecuación logística

## Definición de la ecuación logística 

```python
def Logistica(t, P, delta, K):
    
    dN = delta * P * (1-(P/K))
    
    return dN

dN = Logistica(0, 50, 0.01, 105)
dN
```
